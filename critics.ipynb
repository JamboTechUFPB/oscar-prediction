# Critics Best Actress 

```python
CRITICS_CHOICE_url = "https://en.wikipedia.org/wiki/Critics%27_Choice_Movie_Award_for_Best_Actress"
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

soup = BeautifulSoup(requests.get(CRITICS_CHOICE_url).text, 'html.parser')
tables = soup.find_all('table', class_='wikitable')
```

```python
def isWinner(row):
    return bool(row.find('td',{'style': 'background:#B0C4DE;'})) 
```

```python
#pega todas as vencedoras
data = []

for table in tables:
    rows = table.find_all('tr')
    
    for row in rows[1:]:
        cells = row.find_all('td')
        
        if len(cells) >= 4:
            ano = cells[0].text.strip()
            nome = cells[1].text.strip()
            personagem = cells[2].text.strip()
            filme = cells[3].text.strip()
            
            ganhou = isWinner(row)
            
            data.append([ano[:4], nome, personagem, filme, ganhou])

df_vencedoras = pd.DataFrame(data, columns=['ano', 'nome', 'personagem', 'filme', 'ganhou'])
```

```python
df = pd.DataFrame(columns=['ano', 'nome', 'personagem', 'filme', 'ganhou'])

for table in tables:
    rows = table.find_all('tr')
    current_year = None  
    current_filme = None 

    
    for row in rows[1:]:
    
        ano_th = row.find('th', {'style': 'text-align:center;'})
        if not ano_th:
            ano_th = row.find('th')

        if ano_th is not None and ano_th.find('a'):
            current_year = ano_th.find('a').text.strip()

        ano_atual = current_year

        tds = row.find_all('td')
        if len(tds) >= 3:  
            nome = tds[0].find('a').text if tds[0].find('a') else tds[0].text.strip()
            personagem = tds[1].find('a').text if tds[1].find('a') else tds[1].text.strip()

            if len(tds) > 2:
                filme = tds[2].find('a').text if tds[2].find('a') else tds[2].text.strip()
                if filme:
                    current_filme = filme 
                else:
                    filme = current_filme  
                    
            ganhou = isWinner(row)

            df = pd.concat(
                [
                    df,
                    pd.DataFrame(
                        {
                            'ano': [ano_atual],
                            'nome': [nome],
                            'personagem': [personagem],
                            'filme': [filme],
                            'ganhou': [ganhou]
                        }
                    )
                ],
                ignore_index=True
            )
```
