# Critics Best Actress 

```python
CRITICS_CHOICE_url = "https://en.wikipedia.org/wiki/Critics%27_Choice_Movie_Award_for_Best_Actress"
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

soup = BeautifulSoup(requests.get(CRITICS_CHOICE_url).text, 'html.parser')
tables = soup.find_all('table', class_='wikitable')
```

```python
def isWinner(row):
    return bool(row.find('td',{'style': 'background:#B0C4DE;'})) 
```

```python
df = pd.DataFrame(columns=['ano','nome','personagem','filme','ganhou'])
for table in tables:
    rows = table.find_all('tr')
    current_year = None
    current_filme = None
    for row in rows[1:]:
        tds = row.find_all('td')
        print('tam do tds',len(tds))
        #year
        ano = tds[0].find('a').text
        #print(f"Ano da linha: {ano}")
        if len(ano) == 4:
            current_year = ano
        else:
            ano = current_year
        #print(f"Ano atual da linha: {ano}")
        
        #name

        if len(tds) == 4:
            nome = tds[1].find('a').text
            personagem = tds[2].text
            filme = tds[3].text
            ganhou = isWinner(row)
        else:
            nome = tds[0].find('a').text
            personagem = tds[1].text
            filme = tds[2].text
        ganhou = isWinner(row)

        print(f"Ano: {ano}, Nome: {nome}, Personagem: {personagem}, Filme: {filme}, Ganhou: {ganhou}")

        df = pd.concat(
                            [
                                df,
                                pd.DataFrame(
                                    {
                                        'ano': [current_year],
                                        'nome': [nome],
                                        'personagem': [personagem],
                                        'filme': [filme],
                                        'ganhou': [ganhou]
                                    }
                                )
                            ]
                        )
```
