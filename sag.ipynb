# SAG Best Actress

```python
SAG_AWARDS_url = "https://en.wikipedia.org/wiki/Screen_Actors_Guild_Award_for_Outstanding_Performance_by_a_Female_Actor_in_a_Leading_Role"
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

soup = BeautifulSoup(requests.get(SAG_AWARDS_url).text, 'html.parser')
tables = soup.find_all('table', class_='wikitable sortable')
```

```python
def isWinner(row):
    return bool(row.find('td',{'style': 'background:#FAEB86;'})) 
```

```python
import pandas as pd

dados = [] 

for table in tables:
    rows = table.find_all('tr')
    current_year = None 
    current_filme = None 
    
    for row in rows[1:]:  
        #pula o cabeçalho
        ano_th = row.find('th', {'style': 'text-align:center;'}) or row.find('th')
        
        # Se encontrou um th com o estilo, atualiza o ano atual
        if ano_th and ano_th.find('a'):
            current_year = ano_th.find('a').text     
        
        #usa o ano atual para esta linha
        ano_atual = current_year

        tds = row.find_all('td')
        if not tds or len(tds) < 2:
            #pula linhas inválidas
            continue  
        
        nome = tds[0].find('a').text if tds[0].find('a') else tds[0].text.strip()
        personagem = tds[1].find('a').text if tds[1].find('a') else tds[1].text.strip()
        
        filme = None
        if len(tds) > 2:
            filme = tds[2].find('a').text if tds[2].find('a') else tds[2].text.strip()
            if filme:
                current_filme = filme
            else:
                filme = current_filme  

        #a primeira atriz de cada ano é a vencedora
        ganhou = isWinner(row)

        #adiciona os dados na lista
        dados.append({
            'ano': ano_atual,
            'nome': nome,
            'personagem': personagem,
            'filme': filme,
            'ganhou': ganhou
        })

# Converte a lista para DataFrame
df = pd.DataFrame(dados)
```
